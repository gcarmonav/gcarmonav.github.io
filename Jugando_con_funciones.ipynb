###Definiciones

Continuas: Una función es continua si su gráfica puede dibujarse de un solo trazo. 

Convexas: Es una función, en la cual si dados dos puntos cualesquiera M y N de su gráfica, el segmento que los une queda por arriba de la curva de la función. 

Discontinuas: es aquella función que no es continua, es decir, presenta algún punto en el que existe un salto y la gráfica se rompe. 

Suaves: es una curva que no posee puntos angulosos, por ejemplo: el círculo, la elipse, la parábola, entre otros.

1.1. $f(x) = x^3 + x^2$, x E R

```python
x, y = symbols('x y', integer = True)
```

```python
# Definir la función
y = x**3 + x**2
y
```

```python
# Cálculo de la derivada
diff(y)
```

```python
# Cálculo de la 2da derivada
diff(diff(y))
```

```python
# Definir los puntos donde queremos evaluar nuestra función y su derivada
puntos = np.linspace(-100,100,1000)
```

```python
# Evaluar la función 
f1 = lambdify(x,y, 'numpy')        #valores float
f1(puntos)
```

```python
# Evaluar la derivada
f2 = lambdify(x, diff(y), 'numpy')    # valores float
f2(puntos)
```

```python
# Evaluar la segunda derivada
f3 = lambdify(x, diff(diff(y)), 'numpy')    # valores float
f3(puntos)
```

```python
plt.figure(figsize = (11,6))
plt.plot(puntos, f1(puntos), '-k.', label = r'$f(x) = x^3 + x^2$')
plt.plot(puntos, f2(puntos), '-r.', label = r'$\frac{df(x)}{dx} = 3x^2 + 2x$')
plt.plot(puntos, f3(puntos), '-g.', label = r'$\frac{df^2(x)}{dx^2}  = 6x + 2$')
plt.xlabel('x', fontsize = 16)
plt.ylabel('y', fontsize = 16)
plt.title('Gráfica de función, primera y segunda derivada', fontsize = 14)
plt.xlim(-10,10)
plt.ylim(-100,100)
plt.legend(loc='best', fontsize = 14)
plt.grid()
plt.show()
```

De acuerdo con la gráfica, la función $f(x) = x^3 + x^2$ corresponde a una curva **convexa, suave y continua**

###2.1. Función de Rosenbrock:

imagen.png

```python
def f(X, Y):
    return (100 * ( Y - X ** 2)**2) + (1 - X) ** 2 # se ajustó de acuerdo a la función original
```

```python
#
# Generación de un vector de puntos para x
#
X = np.linspace(start=-2.048, stop=2.048, num=100)
```

```python
#
# Generación de un vector de puntos para y
#
Y = np.linspace(start=-2.048, stop=2.048, num=100)
```

```python
#
# Generación de una matriz para representar
# una malla de puntos
#
X, Y = np.meshgrid(X, Y)
```

```python
#
# Generación de una matriz con los valores de f(x, y)
#
Z = f(X, Y)
```

**Gráfica de la función**

```python

fig = plt.figure(figsize=(10, 10))
ax = fig.gca(projection="3d")
ax.plot_surface(X, Y, Z, cmap=cm.coolwarm, linewidth=1, antialiased=False)
plt.show()
```

**Contorno de la función**

```python
#
#  El punto rojo señala el punto de mínima
#  de la función
#
def plot_contour():
    fig, ax = plt.subplots(figsize=(7, 7))
    ax.set_aspect("equal", "box")
    ax.contour(X, Y, Z)
    ax.grid()
    plt.plot([1], [1], "o", color="red")


plot_contour()
```
